## Stratified Split Based on "Type of Reagent"

On the publication, the author especifies that the splitting of the data into training and test set should ensure that:

1. The data from different **types of reagents** is propotional in both sets.
2. The combination **solvent-FPD** values for a particular molecule stays togethers either in the training or test set.

```python
# Initialize the train ratio
train_ratio = 0.85

# Initialize empty DataFrames train and test data
train_data = pd.DataFrame(columns=data_to_split.columns)
test_data = pd.DataFrame(columns=data_to_split.columns)

# Group by ' Type of Reagent'
for reagent_type, reagent_type_group in data_to_split.groupby(' Type of Reagent'):

    # List of unique reagents in this group
    unique_reagents = reagent_type_group['Name of the Reagent'].unique()

    # Calculate the number of reagents to sample for training
    num_train_reagents = int(len(unique_reagents) * train_ratio)

    # Randomly sample reagents for the training set
    sampled_reagents = np.random.choice(unique_reagents, size=num_train_reagents, replace=False)

    # Add the corresponding data for these reagents to the training set
    train_data = pd.concat([train_data, reagent_type_group[reagent_type_group['Name of the Reagent'].isin(sampled_reagents)]])

    # Add the remaining data to the test set
    test_data = pd.concat([test_data, reagent_type_group[~reagent_type_group['Name of the Reagent'].isin(sampled_reagents)]])

# Reset index for clarity
train_data.reset_index(drop=True, inplace=True)
test_data.reset_index(drop=True, inplace=True)

train_data.shape, test_data.shape  # Display the shape of train and test data
```

```python
# Analyze the proportionality of "Type of Reagent"

# Proportions in the entire dataset
overall_proportions = data_to_split[' Type of Reagent'].value_counts(normalize=True)

# Proportions in the training set
train_proportions = train_data[' Type of Reagent'].value_counts(normalize=True)

# Proportions in the test set
test_proportions = test_data[' Type of Reagent'].value_counts(normalize=True)

# Combine the proportions into a single DataFrame
proportions_df = pd.DataFrame({
    'Overall': overall_proportions,
    'Train': train_proportions,
    'Test': test_proportions
})

proportions_df
```

The proportions of the types of reagents is similar in the three datasets, the initial one containing the 260 molecules (130 x 2 solvents), the training and test set.

```python
# Verify that "Name of the Reagent" values are not split between train and test sets

# Unique reagents in the training list
train_reagents = list(train_data['Name of the Reagent'].unique())

# Unique reagents in the test list
test_reagents = list(test_data['Name of the Reagent'].unique())

# Find reagents that might appear in both sets
common_reagents = [value for value in train_reagents if value in test_reagents]

common_reagents
```

We verified that the same molecule in both solvents got sorted together to either the test or the training set.
